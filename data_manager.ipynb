```python
from PIL import Image
import os
import tensorflow as tf
import random
import numpy as np
import warnings
```

```python
warnings.filterwarnings('ignore')

# Seeding for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)
```

```python
# Paths 
data_dir = "../data/Face Mask Dataset"
train_dir = os.path.join(data_dir, "Train")
val_dir = os.path.join(data_dir, "Validation")
test_dir = os.path.join(data_dir, "Test")
```

```python
# Check corrupted images
def check_images(folder):
    for cls in os.listdir(folder):
        path = os.path.join(folder, cls)
        for img in os.listdir(path):
            try:
                Image.open(os.path.join(path, img))
            except:
                print("Corrupted:", img)
    print(f"{folder} --> Everything Good!")

# Run check
check_images(train_dir)
check_images(val_dir)
check_images(test_dir)
```

```python
#  Count images per class
def count_images(folder):
    for cls in os.listdir(folder):
        path = os.path.join(folder, cls)
        print(f"{cls}: {len(os.listdir(path))}")

print("Train:")
count_images(train_dir)

print("\nValidation:")
count_images(val_dir)

print("\nTest:")
count_images(test_dir)
```
